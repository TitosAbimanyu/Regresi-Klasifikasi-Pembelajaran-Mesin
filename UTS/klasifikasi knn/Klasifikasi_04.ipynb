## preprocessing

```python
from sklearn.preprocessing import LabelEncoder
for col in ['job','marital','education','default','housing', 'loan', 'contact','day','month','poutcome','y']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    data
    print (data['job'],data['marital'],data['education'],data['default'],data['housing'],data['loan'],data['contact'],data['day'],data['month'],data['poutcome'],data['y'])
```

```python
data
```

## Menentukan data x dan y

```python
x = data.drop('y', axis = 1)
y = data['y']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state =0)
```

## Melakukan klasifikasi

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors =1)
```

```python
knn.fit(x_train, y_train)
```

```python
print("Akurasi score sebesar: {}".format(knn.score(x_test,y_test)*100))
```
